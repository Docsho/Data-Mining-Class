```python
from itertools import combinations

transactions = [
    {'A', 'B', 'C'},  
    {'A', 'C'},        
    {'A', 'D'},        
    {'B', 'E', 'F'}  
]

min_support = 0.5
min_count = min_support * len(transactions)
```

```python
def get_frequent_itemsets(candidates, transactions, min_count):
    freq = {}
    for t in transactions:
        for c in candidates:
            if c.issubset(t):
                freq[c] = freq.get(c, 0) + 1
    return {itemset: count for itemset, count in freq.items() if count >= min_count}
```

```python
def apriori(transactions, min_support):
    min_count = min_support * len(transactions)
    items = {frozenset([i]) for t in transactions for i in t}
    L = []
    Lk = get_frequent_itemsets(items, transactions, min_count)
    k = 1
    
    while Lk:
        L.append(Lk)
        candidates = set([
            i.union(j) for i in Lk for j in Lk
            if len(i.union(j)) == k + 1
        ])
        Lk = get_frequent_itemsets(candidates, transactions, min_count)
        k += 1
    
    all_frequent = {}
    for Lk in L:
        all_frequent.update(Lk)
    
    return all_frequent

frequent_itemsets = apriori(transactions, min_support)
```

```python
print("Frequent itemsets (with counts):")
for itemset, count in frequent_itemsets.items():
    print(set(itemset), "->", count)
```
